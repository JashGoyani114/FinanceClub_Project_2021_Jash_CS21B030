```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import matlabpyplot as plot
```

```python
data=pd.read_csv("data.csv")
```

```python
def rma(data,days):
    avg=[]
    sum=0
    for i in range(len(data)):
        if(i<days):
            avg.append(0)
            sum=sum+data[i]
        elif(i==days):
            sum=sum+data[i]
            avg.append(sum/days)
        else:
            tmp=avg[i-1]*(days-1)+data[i]
            avg.append(tmp/days)
    return avg
                                                                                                                                                                                        
```

```python
def TR(data):
    mx=[]
    mx.append(0)
    for i in range(1,len(data)):
        mx.append(max({abs(data['High'][i]-data['Low'][i]),abs(data['High'][i]-data['Low'][i-1]),abs(data['Low'][i]-data['Low'][i-1])}))
    return mx
```

```python
def FCI(AvgLoss,AvgGain,ATRShort,ATRLong,days):
    arr=[]
    for i in range(len(data)-1):
        if(i<days):
            arr.append(0)
        else:
            arr.append((100-100/(1+((ATRShort[i]*AvgGain[i])/(ATRLong[i]*AvgLoss[i])))))
    return arr
```

```python
def RSI(AvgLoss,AvgGain,days):
    arr=[]
    for i in range(len(AvgLoss)-1):
        if(i<days):
            arr.append(0)
        else:
            arr.append((100-100/(1+AvgGain[i]/AvgLoss[i])))
    return arr
    
```

```python
change=[]
gain=[]
loss=[]
change.append(0);
for i in range(1,366):
    change.append(data['Close'][i]-data['Close'][i-1])
    gain.append(max(0,change[i]))
    loss.append(min(0,change[i])*-1)
# display(gain)
                

avgGain = rma(gain, 14)
avgLoss = rma(loss, 14)
TR=TR(data)
ATR_Short=rma(TR,7)
ATR_Long=rma(TR,21)
```

```python
# display(avgGain)
plt.plot(avgGain,label='avgGain')
plt.plot(avgLoss,label='avgLoss')
plt.plot(ATR_Short,label='ATR_Short')
plt.plot(ATR_Long,label='ATR_Long')
plt.xlabel('Time')
plt.ylabel('Avg Gain')
plt.title('Avg Gain')
plt.legend()
```

```python
FCI=FCI(avgLoss,avgGain,ATR_Short,ATR_Long,21)
```

```python
RSI=RSI(avgLoss,avgGain,14)
```

```python
plt.plot(FCI,label='FCI')
plt.plot(RSI,label='RSI')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Indicators')
plt.legend()
```

```python
FCImean= rma(FCI,10)
RSImean= rma(RSI,10)
```

```python
def Signal(FCImean,RSImean):
    arr=[]
    for i in range(len(RSImean)-1):
        arr.append(FCImean[i]>RSImean[i])
    return arr
```

```python
Signal=Signal(FCImean,RSImean)
# display(Signal)
```

```python
plt.plot(FCImean,label='FCImean')
plt.plot(RSImean,label='RSImean')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Indicators')
plt.legend()
```

```python
# initialize variables for backtesting
portfolio_value = 100000
portfolio_shares = 0

# iterate through data and execute trades based on signals
for i in range(1,len(Signal)):
    if Signal[i]>0:
        shares_to_buy = (portfolio_value * 0.1) / data['Close'][i]
        portfolio_shares += shares_to_buy
        portfolio_value -= shares_to_buy * data['Close'][i]
    else:
        portfolio_value += portfolio_shares * data['Close'][i]
        portfolio_shares = 0

# print final portfolio value and Share
print("Final Portfolio Value: $" + str(portfolio_value))
print("Final Portfolio Share: " + str(portfolio_shares))
```

```python
#Q3) IMPLEMENTING A BASIC STRATEGY

ATR=rma(TR,14)

def ATR_RSI_strategy(ATR,RSI, threshold=50):
    # Buy signal
    buy=[]
    sell=[]
    for i in range(len(RSI)):
        buy.append((RSI[i] > threshold) & (ATR[i] > ATR[i-1]))
        # Sell signal
        sell.append((RSI[i] < threshold) & (ATR[i] < ATR[i-1]))
    
    return buy,sell
# Load data and run the strategy

buy,sell = ATR_RSI_strategy(ATR,RSI)
#display(buy)
#display(sell)
```

```python
# initialize variables for backtesting
portfolio_value = 100000
portfolio_shares = 0

# iterate through data and execute trades based on signals
for i in range(1,len(buy)):
    if buy[i]:
        shares_to_buy = (portfolio_value * 0.1) / data['Close'][i]
        portfolio_shares += shares_to_buy
        portfolio_value -= shares_to_buy * data['Close'][i]
    elif sell[i]:
        portfolio_value += portfolio_shares * data['Close'][i]
        portfolio_shares = 0

# print final portfolio value and Share
print("Final Portfolio Value: $" + str(portfolio_value))
print("Final Portfolio Share: " + str(portfolio_shares*data['Close'][len(data)-1]))
```
